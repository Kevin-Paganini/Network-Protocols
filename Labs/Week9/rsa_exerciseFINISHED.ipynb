```python
import random
import math

def extended_euclidean_algorithm(a, b):
    """
    extended_euclidean_algorithm(a, b)

    The result is the largest common divisor for a and b.

    :param a: integer number
    :param b: integer number
    :return:  the largest common divisor for a and b
    """
    if a == 0:
        return b, 0, 1
    else:
        g, y, x = extended_euclidean_algorithm(b % a, a)
        return g, x - (b // a) * y, y


def modular_inverse(e, t):
    """
    modular_inverse(e, t)

    Counts modular multiplicative inverse for e and t.

    :param e: in this case e is a public key exponent
    :param t: and t is an Euler function
    :return:  the result of modular multiplicative inverse for e and t
    """
    g, x, y = extended_euclidean_algorithm(e, t)

    if g != 1:
        raise Exception('Modular inverse does not exist')
    else:
        return x % t
```

```python
def generate_primes(low, high):
    assert low >= 1
    assert high > low
    primes = []
    for x in range(low, high):
        for y in range(2,x):
            if x%y==0:break
        else:
            #print (x,sep=' ', end=' ')
            primes.append(x)
    print(primes)
    return primes
            
    # return a list of prime numbers from low to high
    # remember a number (n) is prime if for all numbers (x) less than n, n % x != 0

def pow_mod(n, x, m):
    return pow(n, x, m)
    
    # This function should return n^x % m, for rsa n^x could end up being much bigger than a 64 bit integer
    # Thus we need to break it up by remembering that:
    # (n*n)%m * (n*n)%m * (n*n)%m * ...do this x times... * (n*n)%m == n^x % m

def encrypt(x, e, n):
    return pow_mod(x, e, n)

def decrypt(x, d, n):
    return pow_mod(x, d, n)
```

```python
generate_primes(1, 1025)
```

```python
primes = generate_primes(1, 1024)
print(primes)

# This is one of the public prime numbers
e = 65537

# The following generate our private numbers
p = random.choice(primes)
q = random.choice(primes)
z = (p-1)*(q-1) # Also called t or the totient
# e and z must be co-prime - gcd(e, z) = 1
while math.gcd(e, z) != 1:
    p = random.choice(primes)
    q = random.choice(primes)
    z = (p-1)*(q-1)
# now we need to compute d
# e*d = 1 (mod z)
d = modular_inverse(e, z)
print(d)

# This is the other public number
n = p*q

text = int(input("Enter the text to be encrypted: (it has to be a number)"))
print("Original text is: ", text)
cypher_text = encrypt(text, e, n)
print("Cypher text is: ", cypher_text)
text = decrypt(cypher_text, d, n)
print("Original text is: ", text)
```
